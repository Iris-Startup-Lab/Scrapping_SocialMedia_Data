```python
def create_social_media_pdf(df, output_path="social_media_report.pdf"):
    """
    Generates a PDF report from a social media DataFrame.

    Args:
        df (pd.DataFrame): The DataFrame containing social media data.
                            Must have columns: 'social_media', 'comment_es',
                            'sentiment', 'emotion', 'categories', 'gender', 'app'.
        output_path (str): The path to save the generated PDF file.
    """
    pdf = FPDF('P', 'mm', 'A4') # Portrait, millimeters, A4 size
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()

    font_family = 'Arial'  # Default font family
    noto_font_path = 'E:/Users/1167486/Local/fonts/Noto_Sans/static/NotoSans-Regular.ttf'
    
    if not os.path.exists(noto_font_path):
        print(f"ERROR: Noto Sans font file not found at '{noto_font_path}'.")
        print("Falling back to Arial. Unicode characters may not render correctly.")
    else:
        try:    
            pdf.add_font('NotoSans', '', noto_font_path)
            font_family = 'NotoSans'
            print(f"Successfully loaded font: Noto Sans from {noto_font_path}")
        except Exception as e:
            print(f"Warning: Could not load/process Noto Sans '{noto_font_path}' ({type(e).__name__}: {e}).")
            print("Falling back to Arial. Unicode characters may not render correctly.")

    # Set font for the title
    pdf.set_font(font_family, '', 16)
    pdf.cell(0, 10, 'Social Media Data Report', 0, 1, 'C')
    pdf.ln(10) # Add some space

    # Set font for table headers
    pdf.set_font(font_family, '', 10)
    
    # Define column headers and widths (in mm)
    # Adjust widths based on your content and page size (A4 is ~210mm wide)
    # Total width should be less than page width minus margins (210 - 15*2 = 180mm)
    col_widths = {
        'Red Social': 25, # Ajustado para que quepa mejor en A4
        'Comment': 75,   # Ajustado para que quepa mejor en A4
        'Sentiment': 20,
        'Emotion': 20,
        'Category': 20,
        'Gender': 20, # Renombrado de 'Genero' a 'Gender' para consistencia con tu df.get()
        'App': 20 # Renombrado de 'Aplicacion' a 'App' para consistencia con tu df.get()
    }
    headers = list(col_widths.keys())

    # Draw table headers
    for header in headers:
        pdf.cell(col_widths[header], 10, header, 1, 0, 'C')
    pdf.ln() # Move to the next line after headers

    # Set font for table data
    pdf.set_font(font_family, '', 9)

    # Iterate over DataFrame rows
    for index, row in df.iterrows():
        # Ensure a new page if content exceeds the current one
        if pdf.get_y() + 20 > pdf.page_break_trigger: # 20 is a safe margin for a new row
            pdf.add_page()
            # Redraw headers on new page
            pdf.set_font(font_family, '', 10)
            for header in headers:
                pdf.cell(col_widths[header], 10, header, 1, 0, 'C')
            pdf.ln()
            pdf.set_font(font_family, '', 9)

        row_start_y = pdf.get_y()
        max_row_height = 10 # Minimum height for a row

        # Prepare text for multi_cell and calculate necessary height for the row
        comment_text = sanitize_text_for_pdf(row.get('comment_es', ''))
        
        # Calculate height needed for the comment cell
        # We temporarily set the X to the comment column's start to simulate text wrapping
        # and get the height
        pdf.set_x(pdf.l_margin + col_widths['Red Social'])
        # split_only=True returns a list of lines, we use its length
        lines_for_comment = pdf.multi_cell(
            w=col_widths['Comment'],
            h=5, # Line height for calculation
            txt=comment_text,
            border=0, # No border for calculation
            align='L',
            split_only=True # Just get the lines, don't draw
        )
        comment_cell_calculated_height = len(lines_for_comment) * 5 # 5mm per line
        max_row_height = max(max_row_height, comment_cell_calculated_height)

        # Now, draw all cells for the current row using the determined max_row_height
        # We need to explicitly manage X and Y positions for each cell
        current_x = pdf.l_margin
        
        # Red Social
        social_media_text = sanitize_text_for_pdf(row.get('social_media', ''))
        pdf.set_xy(current_x, row_start_y)
        pdf.multi_cell(w=col_widths['Red Social'], h=max_row_height, txt=social_media_text, border=1, align='L', fill=False)
        current_x += col_widths['Red Social']

        # Comment
        pdf.set_xy(current_x, row_start_y)
        pdf.multi_cell(w=col_widths['Comment'], h=max_row_height, txt=comment_text, border=1, align='L', fill=False)
        current_x += col_widths['Comment']

        # Sentiment
        sentiment_text = sanitize_text_for_pdf(row.get('sentiment', ''))
        pdf.set_xy(current_x, row_start_y)
        pdf.multi_cell(w=col_widths['Sentiment'], h=max_row_height, txt=sentiment_text, border=1, align='C', fill=False)
        current_x += col_widths['Sentiment']

        # Emotion
        emotion_text = sanitize_text_for_pdf(row.get('emotion', ''))
        pdf.set_xy(current_x, row_start_y)
        pdf.multi_cell(w=col_widths['Emotion'], h=max_row_height, txt=emotion_text, border=1, align='C', fill=False)
        current_x += col_widths['Emotion']

        # Category
        category_text = sanitize_text_for_pdf(row.get('categories', ''))
        pdf.set_xy(current_x, row_start_y)
        pdf.multi_cell(w=col_widths['Category'], h=max_row_height, txt=category_text, border=1, align='C', fill=False)
        current_x += col_widths['Category']

        # Gender
        gender_text = sanitize_text_for_pdf(row.get('gender', ''))
        pdf.set_xy(current_x, row_start_y)
        pdf.multi_cell(w=col_widths['Gender'], h=max_row_height, txt=gender_text, border=1, align='C', fill=False)
        current_x += col_widths['Gender']

        # App
        app_text = sanitize_text_for_pdf(row.get('app', ''))
        pdf.set_xy(current_x, row_start_y)
        pdf.multi_cell(w=col_widths['App'], h=max_row_height, txt=app_text, border=1, align='C', fill=False)
        
        # Move to the start of the next line, using the max_row_height
        pdf.set_xy(pdf.l_margin, row_start_y + max_row_height)

    # Output the PDF
    pdf.output(output_path)
    print(f"PDF report saved to {output_path}")
```

```python
import unicodedata

def sanitize_text_for_pdf(text_input):
    """
    Sanitizes and normalizes text to be more PDF-friendly with Unicode fonts.
    """
    if pd.isna(text_input):
        return ''  # Replace NaN/None with an empty string

    text = str(text_input) # Ensure it's a string

    # Normalize Unicode characters. NFKC is a good general-purpose form.
    try:
        text = unicodedata.normalize('NFKC', text)
    except TypeError:
        pass 

    return text
```
